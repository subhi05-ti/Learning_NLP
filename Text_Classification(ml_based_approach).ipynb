**Lowercasing**

```python
df['review'] = df['review'].str.lower()
```

```python
df['review'][1]
```

**Removing HTML Tags**

```python
import re
def remove_tag(text):
  pattern = re.compile('<.*?>')
  return pattern.sub(r'',text)
```

```python
df['review'] = df['review'].apply(remove_tag)
```

```python
df['review'][1]
```

**Removing Stopwords**

```python
import nltk
nltk.download('punkt')
from nltk.corpus import stopwords
nltk.download('stopwords')
```

```python
stopword = stopwords.words('english')
```

```python
stopword
```

```python
def remove_stopword(text):
  new_text = []
  for w in text.split():
    if w in stopword:
      new_text.append('')
    else:
      new_text.append(w)

  return " ".join(new_text)
```

```python
df['review'] = df['review'].apply(remove_stopword)
```

```python
df['review'][1]
```

```python
# fetature separation
x = df.drop(columns = 'sentiment',axis=1)
y = df['sentiment']
```

```python
x
```

```python
y
```

**Applying lable encoding on y**

```python
from sklearn.preprocessing import LabelEncoder
encoder = LabelEncoder()
y = encoder.fit_transform(y)
```

```python
y
```

**Train_Test_split**

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test = train_test_split(x,y,test_size=0.2,random_state=1)
```

```python
print(x.shape,x_train.shape,x_test.shape)
```

#Applying BOW

```python
from sklearn.feature_extraction.text import CountVectorizer
```

```python
cv = CountVectorizer()
x_train_bow = cv.fit_transform(x_train['review']).toarray()
x_test_bow = cv.transform(x_test['review']).toarray()
```

```python
x_train_bow.shape
```

**Training model on GaussianNB**

```python
from sklearn.naive_bayes import GaussianNB
from sklearn.metrics import accuracy_score,confusion_matrix
```

```python
gnb = GaussianNB()
gnb.fit(x_train_bow,y_train)
```

```python
y_pred = gnb.predict(x_test_bow)
accuracy_score(y_test,y_pred)
```

```python
confusion_matrix(y_test,y_pred)
```

**Training on RandomForest**

```python
from sklearn.ensemble import RandomForestClassifier
```

```python
rf = RandomForestClassifier()
rf.fit(x_train_bow,y_train)
```

```python
y_pred2 = rf.predict(x_test_bow)
accuracy_score(y_test,y_pred2)
```

#Applying Tfidf

```python
from sklearn.feature_extraction.text import TfidfVectorizer
```

```python
tfidf = TfidfVectorizer()
```

```python
x_train_tfidf = tfidf.fit_transform(x_train['review']).toarray()
x_test_tfidf = tfidf.transform(x_test['review']).toarray()
```

```python
rf.fit(x_train_tfidf,y_train)
y_pred3 = rf.predict(x_test_tfidf)
accuracy_score(y_test,y_pred3)
```
