**Converting numpy arrays to pytorch tensors**

```python
x_train_ten = torch.from_numpy(x_train)
x_test_ten = torch.from_numpy(x_test)
y_train_ten = torch.from_numpy(y_train)
y_test_ten = torch.from_numpy(y_test)
```

```python
print(x_train_ten.shape)
print(y_train_ten.shape)
```

**Defining model**

```python
class MysimpleNN():
  def __init__(self,x):
    self.weights = torch.rand(x.shape[1],1,dtype=torch.float64,requires_grad=True)
    self.bias = torch.zeros(1,dtype=torch.float64,requires_grad=True)

  def forward(self,x):
    z = torch.matmul(x,self.weights) + self.bias
    y_pred = torch.sigmoid(z)
    return y_pred

  def loss(self,y_pred,y):
    epsilon = 1e-7
    y_pred = torch.clamp(y_pred,epsilon,1-epsilon)

    # calculate loss
    loss = -(y_train_ten + torch.log(y_pred) + (1 - y_train_ten) + torch.log(1-y_pred)).mean()
    return loss
```

**Important Parameters**

```python
learning_rate = 0.1
epochs = 25
```

**Training Pipeline**

```python
# create model
model = MysimpleNN(x_train_ten)

# define loop
for epoch in range(epochs):

  # forward pass
  y_pred = model.forward(x_train_ten)

  # loss
  loss = model.loss(y_pred,y_train_ten)

  # backward pass
  loss.backward()

  # parameters update
  with torch.no_grad():
    model.weights -= learning_rate * model.weights.grad
    model.bias -= learning_rate * model.bias.grad

  # zero gradient / cleaning grad
  model.weights.grad.zero_()
  model.bias.grad.zero_()

  # print loss in each epoch
  print(f"epoch:{epoch +1 }, Loss : {loss.item()}")
```

**Evaluation**

```python
with torch.no_grad():
  y_pred = model.forward(x_test_ten)
  y_pred = (y_pred > 0.5).float()
  accuracy = (y_pred == y_test_ten).float().mean()
print(f"Accuracy : {accuracy.item()}")

#print(y_pred)
```
