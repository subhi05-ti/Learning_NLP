```python
from sklearn.datasets import make_classification
import torch
```

```python
# making synthetic classification dataset
x,y = make_classification(
    n_samples = 10,          # number of sample
    n_features = 2,         # number of features
    n_informative = 2,      # number of informative features
    n_redundant = 0,        # number of redundant features
    n_classes = 2,          # nnumber of classes
    random_state = 42       # for reproducibility
)
```

```python
# converting the data to pytorch tensor
x = torch.tensor(x,dtype=torch.float32)
y = torch.tensor(y,dtype=torch.float32)
```

```python
# importing the dataset and dataloader class
from torch.utils.data import Dataset, DataLoader
```

```python
# creating class
class CustomDataset(Dataset):
  def __init__(self,features,labels):
    self.features = features
    self.labels = labels

  def __len__(self):
    return self.features.shape[0]

  def __getitem__(self,index):
    return self.features[index],self.labels[index]
```

```python
dataset = CustomDataset(x,y)
```

```python
len(dataset)
```

```python
dataset[6]
```

```python
# creating instance of Dataloader class
dataloader = DataLoader(dataset,batch_size=2,shuffle=False)
```

```python
for batch_features,batch_labels in dataloader:
  print(batch_features)
  print(batch_labels)
  print("-"*70)
```
