```python
def remove_urls(data):
  text = re.compile(r'https?://\S+|WWW\.\S+')
  return text.sub(r'',data)
```

#Removing Punctuations

```python
import string,time
string.punctuation
```

```python
exclude = string.punctuation
exclude
```

```python
def remove_punc(text):
  for char in exclude:
    text = text.replace(char,'')
  return text
```

```python
text = 'string. With, punctuations?'
print(remove_punc(text))
```

```python
start = time.time()
print(remove_punc(text))
time1 = time.time() - start
print(time1*50000)
# this takes so much of time
```

```python
# time saving technique
def remove_punc1(text):
  return text.translate(str.maketrans('','',exclude))
```

```python
start = time.time()
print(remove_punc1(text))
time2 = time.time() - start
print(time2*50000)
```

```python
time1/time2
# Time2 is 10 times faster then time1
```

#Removing ShortHands

```python
char_words = {
    'AFAIK':'As Far As I Know',
    'AFK':'Away From Keyboard',
    'ASAP':'As Soon As Possible',
    'ATK':'At The Keyboard',
    'ATM':'At The Moment',
    'A3':'Anytime, Anywhere, Anyplace',
    'BAK':'Back At Keyboard',
    'BBL':'Be Back Later',
    'BBS':'Be Back Soon',
    'BFN':'Bye For Now',
    'B4N':'Bye For Now',
    'BRB':'Be Right Back',
    'BRT':'Be Right There',
    'BTW':'By The Way',
    'B4':'Before',
    'B4N':'Bye For Now',
    'CU':'See You',
    'CUL8R':'See You Later',
    'CYA':'See You',
    'FAQ':'Frequently Asked Questions',
    'FC':'Fingers Crossed',
    'FWIW':'For What It\'s Worth',
    'FYI':'For Your Information',
    'GAL':'Get A Life',
    'GG':'Good Game',
    'GN':'Good Night',
    'GMTA':'Great Minds Think Alike',
    'GR8':'Great!',
    'G9':'Genius',
    'IC':'I See',
    'ICQ':'I Seek you (also a chat program)',
    'ILU':'ILU: I Love You',
    'IMHO':'In My Honest/Humble Opinion',
    'IMO':'In My Opinion',
    'IOW':'In Other Words',
}
```

```python
char_words
```

```python
def chat_conversion(text):
  new_text = []
  for w in text.split():
    if w.upper() in char_words:
      new_text.append(char_words[w.upper()])
    else:
      new_text.append(w)
  return " ".join(new_text)
```

```python
text = 'IMHO he is the best'
chat_conversion(text)
```

```python
text = 'IK'
chat_conversion(text)
```

#Removing Stop Words

```python
from nltk.corpus import stopwords
```

```python
import nltk
nltk.download('stopwords')
stopwords = stopwords.words('english')
stopwords
```

```python
# def remove_stopwords(text):

#   for w in text.split():
#     if w in stopwords:
#       text = text.replace(w,'')
#     else:
#       text
#   return (text)
  ## if we use this type of code then it will replace all occurance of w
```

```python
def remove_stopwords(text):
  new_text = []
  for w in text.split():
    if w in stopwords:
      new_text.append('')
    else:
      new_text.append(w)

  return " ".join(new_text)
```

```python
text = 'If you want, I can help you build a full text-cleaning function remove punctuation, emojis, numbers, etc.'
remove_stopwords(text)
```

#Handling emojis

**1.Removing emojis**

```python
import re
def remoce_emoji(text):
  emoji_text = re.compile("["
    u"\U0001F600-\U0001F64F"  # emotions
    u"\U0001F300-\U0001F5FF"  # symbols and pictographs
    u"\U0001F680-\U0001F6FF"  # transport and map symbols
    u"\U0001F1E0-\U0001F1FF"  # flags (ios)
    u"\U00002702-\U000027B0"  # Dingbats
    u"\U000024C2-\U0001F251"
   "]+", flags=re.UNICODE)
  return emoji_text.sub(r'',text)
```

```python
remoce_emoji('love the movie.It was 😘😘')
```

**5. Stemming**

```python
from nltk.stem.porter import PorterStemmer
```

```python
ps = PorterStemmer()
def stem_word(text):
  return " ".join([ps.stem(w) for w in text.split()])
```

```python
text = "wlak walks walking walked"
stem_word(text)
```

**6. Lammatization**

```python
from nltk.stem import WordNetLemmatizer
wordnet_lemmatizer = WordNetLemmatizer()

sentence = 'He was running and eating at the same time. He has bad habit of swimming after playing long hours in the sun'
punctuations = '?:!.,;'
sentence_word = word_tokenize(sentence) # Corrected function name
for word in sentence_word:
  if word in punctuations:
    sentence_word.remove(word)
sentence_word

# printing style
print("{0:20}{1:20}".format("Word","Lemma"))
for word in sentence_word:
  print("{0:20}{1:20}".format(word,wordnet_lemmatizer.lemmatize(word,pos='v')))
```
