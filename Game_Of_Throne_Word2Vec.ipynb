```python
import gensim
import os
```

```python
# creating folder
os.makedirs('data',exist_ok=True)
```

```python
# getting each file from data folder and for each iteration adding it to the stoty set
import nltk
nltk.download('punkt_tab')
from nltk import sent_tokenize
from gensim.utils import simple_preprocess

story = []
for filename in os.listdir('data'):
  f = open(os.path.join('data',filename), encoding='latin-1')
  corpus = f.read()
  raw_sent = sent_tokenize(corpus)
  for sent in raw_sent:
    story.append(simple_preprocess(sent))
```

```python
story
```

```python
len(story)
```

```python
# creating object
model = gensim.models.Word2Vec(
    window=10,
    min_count=2
)
```

```python
model.build_vocab(story)
```

```python
# training the model
model.train(story,total_examples=model.corpus_count,epochs=model.epochs)
```

```python
# finding most similar words
model.wv.most_similar('daenerys')
```

```python
# finding odd one out
model.wv.doesnt_match(['cersei','jaime','bronn','tyrion'])
```

```python
model.wv.doesnt_match(['jon','rikon','robb','arya','sansa','bran'])
```

```python
# visualizing the vector of a particular word
model.wv['jon']
```

```python
# finding similarity between words
model.wv.similarity('arya','sansa')
```

```python
# finding similarity between words
model.wv.similarity('tywin','sansa')
```
