#Removing HTML Tags

```python
sample_text = '<h1>Admin Map (Dummy)</h1><p>A tiny mockup showing heavily / medium / less crowded markers and a legend.</p><div class="map" id="map"><div class="marker red" style="left: 18%; top: 22%;" data-title="Market Square ‚Äî Heavily crowded">H</div><div class="marker orange" style="left: 44%; top: 36%;" data-title="Central Park ‚Äî Medium crowded">M</div><div class="marker green" style="left: 72%; top: 60%;" data-title="Riverside ‚Äî Less crowded">L</div></div>'
```

```python
sample_text
```

```python
# importing regular expression
import re
def scripthtml(data):
  p = re.compile(r'<.*?>')
  return p.sub('',data)
```

```python
scripthtml(sample_text)
```

```python
sample_text2 = '<style>body {font-family: Arial, sans-serif;margin: 0;padding: 40px;background: #f3f3f3;}</style> <script>function showMessage const output = document.getElementById("outputoutput.innerHTML = "You clicked the button!";let a = 5;let b = 7;console.log("Sum = ", a + b);}</script> '
```

```python
sample_text2
```

```python
def remove(data):
  # removing css
  data = re.sub(r'<style.*?>.*?</style>', '',data, flags=re.DOTALL | re.IGNORECASE)

  # removing script
  data = re.sub(r'<script.*?>.*?</script>', '',data, flags = re.DOTALL | re.IGNORECASE)

  # remmoving html
  #data = re.sub(r'<.*?>', '', data)

  return data
  print(remove(sample_text2))
```

```python
remove(sample_text2)
```
