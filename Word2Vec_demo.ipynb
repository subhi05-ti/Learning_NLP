```python
from zipfile import ZipFile
file_name = 'glove.6B.zip'
with ZipFile(file_name,'r') as zip:
  zip.extractall('temp_folder')
```

```python
model = KeyedVectors.load_word2vec_format('/content/temp_folder/glove.6B.300d.txt', binary=False, no_header=True)
```

```python
# representation of a paticular word in form of vector
model['man']
```

```python
# checking the shape
model['man'].shape
```

```python
# finding most similar word
model.most_similar('man')
```

```python
# finding most similar word
model.most_similar('bomb')
```

```python
# finding cosine similarity between two word
model.similarity('diwali','robber')
```

```python
# finding odd one out
model.doesnt_match(['PHP','java','monkey'])
```

```python
vec = model['king'] - model['man'] + model['woman']
model.most_similar([vec])
```

```python
vec = model['robber'] - model['car'] + model['plane']
model.most_similar([vec])
```
