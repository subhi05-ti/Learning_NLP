```python
import spacy
```

```python
nlp = spacy.load('en_core_web_sm')
```

```python
doc = nlp(u"I will google about facebook")
```

```python
# to show
doc.text
```

```python
doc[2]
```

```python
# coarse grade pos
doc[2].pos_
```

```python
# fine grade pos
doc[2].tag_
```

```python
# to expain the short form
spacy.explain('VB')
```

```python
# printing the pos for each word
for word in doc:
  print(word.text,"-------->",word.pos_,word.tag_,spacy.explain(word.tag_))
```

```python
doc2 = nlp(u"I left the room")
for word in doc2:
  print(word.text,"--------->",word.pos_,word.tag_,spacy.explain(word.tag_))
```

```python
doc2 = nlp(u"To the left of the room")
for word in doc2:
  print(word.text,"--------->",word.pos_,word.tag_,spacy.explain(word.tag_))
```

```python
# spacy also used to visualize
from spacy import displacy
doc4 = nlp(u"The quick brown fox jumped over the lazy dog")
```

```python
displacy.render(doc4,style='dep')
```

```python
options = {
    'distance':80,
    'compact': True,
    'color':'#fff',
    'bg':'#00a65a'
}
```

```python
displacy.render(doc4,style='dep',options = options)
```
