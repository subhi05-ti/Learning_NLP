```python
# importing the dependencies
import numpy as np
import pandas as pd
import os
import gensim
```

```python
import nltk
nltk.download('punkt_tab')
from nltk import sent_tokenize
from gensim.utils import simple_preprocess
```

```python
# creating folder in colab
os.makedirs('data',exist_ok=True)
```

```python
story= []
for filename in os.listdir('data'):
  f = open(os.path.join('data',filename))
  corpus = f.read()
  raw_data = sent_tokenize(corpus)
  for data in raw_data:
    story.append(simple_preprocess(data))
```

```python
story
```

```python
from nltk.corpus import stopwords
nltk.download('stopwords')
stopword = stopwords.words('english')
```

```python
stopword
```

**Removing stopwords**

```python
def remove_stopwords(text):
  new_text = []
  for w in text:
    if w in stopword:
      new_text.append('')

    else:
      new_text.append(w)

  return " ".join(new_text)
```

```python
story = [remove_stopwords(text) for text in story]
```

```python
story
```

```python
story = [
    sentence.split()   # list of tokens
    for sentence in story   # your text broken into sentences
]
```

```python
story
```

```python
# making object of gensim class
model = gensim.models.Word2Vec(
    window =10,
    min_count=2
)
```

```python
model.build_vocab(story)
```

```python
# training the model
model.train(story,total_examples=model.corpus_count,epochs=model.epochs)
```

```python
model.wv.most_similar('monica')
```

```python
model.wv.similarity('ross','monica')
```

```python
model.wv.get_normed_vectors().shape
```

**Reducing the dimensionality**

```python
from sklearn.decomposition import PCA
pca = PCA(n_components=3)
```

```python
X = pca.fit_transform(model.wv.get_normed_vectors())
```
